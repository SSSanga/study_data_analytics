### 목표: 수술시간 
### 설명: 수술기법, 스테로이드치료, Location of herniation, 수술실패여부, Seg Angle(raw)

```python
# 추출
df_ROS_extract= df_ROS[['수술시간', '수술기법', '스테로이드치료', 'Location of herniation', '수술실패여부', 'Seg Angle(raw)']]
df_ROS_extract
```

```python
## 각도를 생각했을 떄 중앙값을 기준으로 -14.5<= x <= 14.5 
condition = (df_ROS['Seg Angle(raw)'] >= -14.500000 ) & (df_ROS['Seg Angle(raw)'] <= 14.500000 )
df_ROS_angle = df_ROS.loc[condition]
df_ROS_angle
```

```python
# df_ROS_extract.pivot_table(index=['스테로이드치료', '수술실패여부', ], columns =['Location of herniation'], values='수술시간', aggfunc=[np.median])
# angle -14.5<= x <= 14.5 인 값들에서 수술시간 value로 두고 확인. 
df_ROS_angle.pivot_table(index=['스테로이드치료', '수술실패여부', ], columns =['Location of herniation'], values='수술시간', aggfunc=[np.median, np.mean, np.std])
```

```python
## -14.5 <= x <= 14.5 외의 각도: x < -14.5 & 14.5< x 
condition = (df_ROS['Seg Angle(raw)'] < -14.500000 ) | (df_ROS['Seg Angle(raw)'] > 14.500000 )
df_ROS_outangle = df_ROS.loc[condition]
```

```python
df_ROS_outangle.pivot_table(index=['스테로이드치료', '수술실패여부', ], columns =['Location of herniation'], values='수술시간', aggfunc=[np.median, np.mean, np.std])
```
